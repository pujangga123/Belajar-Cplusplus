```python
import pandas as pd

df = pd.read_csv("data-kebakaran.csv")
df.head()
```

## _OUTER JOIN_

```python
df1 = df[df.bulan==1].groupby('sumber_info').korban_jiwa.sum()
df1
```

```python
df2 = df[df.bulan==2].groupby('sumber_info').korban_jiwa.sum()
df2
```

operasi _OUTER JOIN_ dilakukan lewat perintah `concat`

```python
df3 = pd.concat([df1,df2], axis=1)

# beri nama kolom
# ref: https://www.geeksforgeeks.org/add-column-names-to-dataframe-in-pandas/

# data yang kosong, isi dengan angka 0
# ref: https://datatofish.com/replace-nan-values-with-zeros/

# tampilkan data
df3
```

## INNER JOIN

```python
df4 = pd.merge(df1, df2, left_index=True, right_index=True)
df4
```
